###### Defining training set and test set with independent variables with correlations above 0.3

```python
X_train=training_df[["At Least Bachelors's Degree",'Graduate Degree','White (Not Latino) Population', 'African American Population','Asian American Population','Total Population','Construction.extraction.maintenance.and.repair.occupations',
'White_Asian','SIRE_homogeneity']]
y_train=training_df['Y']
X_test=merged_df[["At Least Bachelors's Degree",'Graduate Degree','White (Not Latino) Population', 'African American Population','Asian American Population','Total Population','Construction.extraction.maintenance.and.repair.occupations',
'White_Asian','SIRE_homogeneity']].dropna()
```

###### Linear Regressiom

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
prediction=reg.predict(X_test)
prediction
```

###### Arranging results dataframe
'rep_prediction'=Predicted Republican vote Percentage
'dem_prediction'=Predicted Democrat vote Percentage

```python
results=merged_df[['State','County','Votes','votes16_trumpd','votes16_clintonh','trump%', 'hilary%']].dropna().reset_index(drop=True)
predicted=pd.DataFrame(prediction).rename(columns={0:'rep_prediction'})
results_df=pd.concat([results,predicted],axis=1)
results_df['dem_prediction']=100-results_df.rep_prediction
results_df.head()
```

###### Calculating prediction accuracy by county

```python
results_df['rep_win_pred']=0#binary column for predicted republican win
results_df['rep_win_real']=0#binary column for real republican win
results_df['correct_pred']=0#binary column for correct prediction
results_df['rep_pred_votes']=results_df.rep_prediction*results_df.Votes*.01# converting % vote to actual votes
results_df['dem_pred_votes']=results_df.dem_prediction*results_df.Votes*.01# converting % vote to actual votes
results_df.loc[results_df['trump%']>results_df['hilary%'],'rep_win_real']=1#converting column values to 1 for win
results_df.loc[results_df.rep_prediction>results_df.dem_prediction,'rep_win_pred']=1#converting column values to 1 for win
results_df.loc[results_df.rep_win_pred==results_df.rep_win_real, 'correct_pred']=1#converting column values to 1 for correct prediction
results_df['correct_pred'].value_counts()/len(results_df)*100 # calculating the accuracy of prediction.1=correct prediction and 0=incorrect prediction
```

###### Accuracy of 92% achieved when broken down by County

###### Calculating prediction accuracy by state

```python
by_state=results_df.groupby(['State']).agg({'rep_pred_votes':'sum','dem_pred_votes':'sum','votes16_trumpd':'sum',
       'votes16_clintonh':'sum'}).reset_index(drop=True)
combined_state=pd.concat([ec_df,by_state],ignore_index=False,axis=1)
combined_state['pred_ec_votes_rep']=0
combined_state['real_ec_votes_rep']=0
combined_state['correct_pred']=0
combined_state.loc[combined_state['rep_pred_votes']>combined_state['dem_pred_votes'],'pred_ec_votes_rep']=combined_state['EC_votes']
combined_state.loc[combined_state['votes16_trumpd']>combined_state['votes16_clintonh'],'real_ec_votes_rep']=combined_state['EC_votes']
combined_state.loc[combined_state['pred_ec_votes_rep']==combined_state['real_ec_votes_rep'],'correct_pred']=1
combined_state['correct_pred'].value_counts()/len(combined_state)*100
```
