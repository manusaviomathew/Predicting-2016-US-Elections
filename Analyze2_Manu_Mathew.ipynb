###### 2. Would you say the three elections are similar to each other?(Note that this is a subjective question. How do you define similarity here? Tell us how you're defining similarity and based on your definition, tell us whether or not these elections are the similar.)

```python
print(county['per_rep08%'].mean())
print(county['per_rep12%'].mean())
print(county['trump%'].mean())
x=county['per_rep08%']
y=county['per_rep12%']
z=county['trump%']
print(stats.ttest_ind(x,y))
print(stats.ttest_ind(x,z))
print(stats.ttest_ind(y,z))
```

 On superimposing the graphs, it is evident that although the elections are nearly identical and are overlapping, the mean of vote shares have been increasing throughout these elections. In other words, republican party vote share (popularity) is seen increasing throughout the election years. This can be confirmed from the means of each election which are 57.7% in 2008, 60.79% in 2012 and 66.72% in 2016. This can also be statistically proved using t-tests which would give a p value lesser than 0.05 for all 3 election combinations (2008-2012,2008-2016,2012-2016).The null hypothesis here would be that the mean of each election are the same.

###### 2. We are asking you to fit the fraction of votes that the Republican party gets within a county, this value is any continuous value between 0 and 1. What model is appropriate for this sort of problem?

A multiple linear regression can be done to develop a model to fit the fraction of Republican party vote share against the data columns with correlation coefficients greatr than aboslute value of 0.3

###### 3. How good is the fit of your model? (You can fit and evaluate it on the entire dataset.)

```python

#entire dataset

republic_df=refined_df.loc[:,['rep_fraction',"At Least Bachelors's Degree",'Graduate Degree','Construction.extraction.maintenance.and.repair.occupations',
'White', 'Black', 'Asian','SIRE_homogeneity','Children.in.single.parent.households','Unemployment','Total Population','Sexually.transmitted.infections','HIV.prevalence.rate',
'Violent.crime','Homicide.rate' ]]
from statsmodels.regression import linear_model
import statsmodels.api as sm
X = sm.add_constant(republic_df[["At Least Bachelors's Degree",'Graduate Degree','Construction.extraction.maintenance.and.repair.occupations',
'White', 'Black', 'Asian','SIRE_homogeneity','Children.in.single.parent.households','Unemployment','Total Population','Sexually.transmitted.infections','HIV.prevalence.rate',
'Violent.crime','Homicide.rate']].values)
y = republic_df['rep_fraction'].values
model = sm.OLS(y,X,missing='drop').fit()
model.summary()
```

```python
republic_df.isna().sum()
```

The null hypothesis in the above multiple linear regression model is that "there is no correlation between each data columns and republican vote share". The above model suggests there is no correlation between 'Black' racial group with Republican vote share which leads to a Type 2 error. This error can be due to the model trying to fit the entire independent variable datasets although some independent variables do not contribute to the Republican vote share in reality. 

On further analysis, it was noted that columns 'Sexually.transmitted.infections','HIV.prevalence.rate','Violent.crime' and 'Homicide.rate' have more than 6% 'NaN' values hence a probability of minimum 6% that the correlation coefficients are not true representation. This means that there is more than 5% probability that we fail to reject the null hypothesis ('there is no correlation between the data columns and republican vote share'). This can compromise the reliability of the model as the model would try to fit the entire datasets of independent variable.    

A reliable model can be developed by neglecting the independent variables such as 'Sexually.transmitted.infections','HIV.prevalence.rate','Violent.crime' and 'Homicide.rate'

```python
rep_df=refined_df.loc[:,['rep_fraction',"At Least Bachelors's Degree",'Graduate Degree','Construction.extraction.maintenance.and.repair.occupations',
'White', 'Black', 'Asian','SIRE_homogeneity','Children.in.single.parent.households','Unemployment','Total Population' ]]
from statsmodels.regression import linear_model
import statsmodels.api as sm
X = sm.add_constant(rep_df[["At Least Bachelors's Degree",'Graduate Degree','Construction.extraction.maintenance.and.repair.occupations',
'White', 'Black', 'Asian','SIRE_homogeneity','Children.in.single.parent.households','Unemployment','Total Population']].values)
y = rep_df['rep_fraction'].values
model = sm.OLS(y,X,missing='drop').fit()
model.summary()
```

The accuracy of above model has dropped from 70% to 60% but the model is more reliable as all the p value of all independent variables are less than 5% hence they are all correlated.
